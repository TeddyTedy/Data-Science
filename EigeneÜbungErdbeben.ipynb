```python
import matplotlib.pyplot as plt
import seaborn as sns
# import sklearn.datasets
import numpy as np
import pandas as pd


# Set default Seaborn style
sns.set()

# csv einlesen, Seperator nicht notwendig
# Liste der Erdbeben der letzten 8 Monate
# Quelle: https://earthquake.usgs.gov/earthquakes/map/?extent=-89.9976,-268.59375&extent=89.99754,628.59375&range=search&timeZone=utc&search=%7B%22name%22:%22Search%20Results%22,%22params%22:%7B%22starttime%22:%222021-02-12%2000:00:00%22,%22endtime%22:%222021-10-12%2023:59:59%22,%22minmagnitude%22:2.5,%22orderby%22:%22time%22%7D%7D
earthquake=pd.read_csv("Erdbeben.csv")
```

```python
earthquake
```

```python
#alle Location Sources
data = earthquake
list(data.locationSource)
```

```python
#distinct location Sources
distinctLocationSource = []
for x in data.locationSource:
    if x not in distinctLocationSource:
        distinctLocationSource.append(x)
print(distinctLocationSource)
```

```python
#Sort Data by LocationSource
data.sort_values(by=['locationSource'])
```

```python
# Erstelle den Plot
_ = plt.hist(data['mag'])


plt.title('Magnitude')
plt.xlabel('Magnitude')
plt.ylabel('Number of Earthquakes')

# Zeige Plot an
plt.show()
```

```python
#'pr', 'ci', 'ak', 'us', 'nn', 'hv', 'nc', 'tx', 'av', 'mb', 'ok', 'uu', 'se', 'nm', 'uw', 'ew'
#Tabelle aufteilen


groups = data.groupby(data.locationSource)
pr_df = groups.get_group("pr")
ci_df = groups.get_group("ci")
ak_df = groups.get_group("ak")
us_df = groups.get_group("us")
nn_df = groups.get_group("nn")
hv_df = groups.get_group("hv")
nc_df = groups.get_group("nc")
tx_df = groups.get_group("tx")
av_df = groups.get_group("av")
mb_df = groups.get_group("mb")
ok_df = groups.get_group("ok")
uu_df = groups.get_group("uu")
se_df = groups.get_group("se")
nm_df = groups.get_group("nm")
uw_df = groups.get_group("uw")
ew_df = groups.get_group("ew")
#Zu viele Daten um es schön darzustellen
# print("Group with Location: pr")
# print(pr_df,"\n")

# print("Group with Location: ci")
# print(ci_df,"\n")

# print("Group with Location: ak")
# print(ak_df,"\n")
```
